3) Создайте рандомный вектор и отсортируйте его.

```python
%%time

size = 10000

vector = np.random.random(size)

sorted_vector = np.sort(vector)

print("Исходный вектор:\n", vector)
print("Отсортированный вектор:\n", sorted_vector)
```

```python
%%time

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return np.concatenate((quicksort(less), [pivot], quicksort(greater)))

def babblesort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

sorted_vector = quicksort(vector)
print("Исходный вектор:\n", vector)
print("Отсортированный вектор:\n", sorted_vector)
```

7) Реализовать кодирование длин серий (Run-length encoding). Дан вектор x.
Необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит
числа, а второй - сколько раз их нужно повторить. Пример: x = np.array([2, 2, 2, 3, 3, 3, 5]). Ответ: (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
%%time

def run_length_encoding(x):
    change_indexes = np.where(x[:-1] != x[1:])[0] + 1
    concat = np.concatenate(([0], change_indexes, [len(x)]))
    lengths = np.diff(concat)
    values = x[concat[1:] - 1]
    return values, lengths

def unique_count(x):
    return np.unique(x, return_counts=True)

x = np.array([2, 2, 2, 3, 3, 3, 5, 3])

result = run_length_encoding(x)
print("Кодирование длин:\n", result)

result = unique_count(x)
print("Число уникальных элементов:\n", result)
```

```python
%%time

numbers = []
counts = []

current_number = None
current_count = 0

def add_current():
    if current_number is not None:
        numbers.append(current_number)
        counts.append(current_count)

for number in x:
    if number == current_number:
        current_count += 1
    else:
        add_current()
        current_number = number
        current_count = 1
add_current()

print(numbers, counts)
```

8) Написать функцию wipe_even(arr, target_value, in_place), принимающую на вход
массив целых чисел arr, и возвращающую массив, полученный из arr путём замены
всех чётных элементов на target_value. Если target_value не указано, то оно должно считаться равным числу 0. Если указан параметр in_place и он равен True, то функция должна менять исходный массив, а если не указан или указан в False, то сохранять его неизменным.

```python
def wipe_even(arr, target_value=0, in_place=False):
    if not in_place:
        arr = arr.copy()
    arr[np.mod(arr, 2) == 0] = target_value
    return arr

x = np.array([1, 2, 3, 4])
result = wipe_even(x)
print("Зануление в копии:", result)
print("Исходный массив:", x)
wipe_even(x, 9, True)
print("На месте:", x)
```

```python
def wipe_even(arr, target_value=0, in_place=False):
    if not in_place:
        arr = arr.copy()
    for i in range(len(arr)):
        if arr[i] % 2 == 0:
            arr[i] = target_value
    return arr

x = np.array([1, 2, 3, 4])
result = wipe_even(x)
print("Зануление в копии:", result)
print("Исходный массив:", x)
wipe_even(x, 9, True)
print("На месте:", x)
```

10) Найдите N наибольших значений в векторе.

```python
def largest_unique_np(arr, n):
    if n == 0:
      return np.empty((0,))
    return np.sort(np.unique(arr))[-n:]

def largest_duplicates_np(arr, n):
    if n == 0:
      return np.empty((0,))
    return np.sort(arr)[-n:]

def largest_unique(arr, n):
    if n == 0:
      return np.empty((0,))
    largest_unique = set()
    for value in arr:
        if len(largest_unique) < n or value > min(largest_unique) :
            largest_unique.add(value)
            if len(largest_unique) > n:
                largest_unique.remove(min(largest_unique))
    result = list(largest_unique)
    result.sort()
    return result

def largest_duplicates(arr, n):
    if n == 0:
      return np.empty((0,))
    largest = [-np.inf] * min(n, arr.size)
    for value in arr:
        if value > largest[0]:
            largest[0] = value
            largest.sort()
    return largest

N = 10
#vector = np.array([0, 2, 4, 2, 1])
vector = np.random.randint(0, 10001, size=999999)
```

```python
%%time

print("Используя NumPy:")
unique = largest_unique_np(vector, N)
duplicates = largest_duplicates_np(vector, N)

print("С учетом повторений чисел:", unique)
print("Без учета повторений чисел:", duplicates)
```

```python
%%time

print("Своя реализация:")
unique = largest_unique(vector, N)
duplicates = largest_duplicates(vector, N)

print("С учетом повторений чисел:", unique)
print("Без учета повторений чисел:", duplicates)
```
