6) ÐÐ°Ð¹Ñ‚Ð¸ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚ Ð´ÐµÐ»ÐµÐ½Ð¸Ñ Ð¼Ð¸Ð½Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð½Ð° Ð¼Ð°ÐºÑÐ¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ðµ Ð² ÐºÐ°Ð¶Ð´Ð¾Ð¼ ÑÑ‚Ð¾Ð»Ð±Ñ†Ðµ Ð´Ð²ÑƒÐ¼ÐµÑ€Ð½Ð¾Ð³Ð¾ Ð¼Ð°ÑÑÐ¸Ð²Ð°, Ð¸ÑÐ¿Ð¾Ð»ÑŒÐ·ÑƒÑ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ np.apply_along_axis().

```python
arr = np.array([[4, -2, -3],
                [8,  3, -6],
                [1, 10,  0]])

def divide_min_max(column):
    min_val = np.min(column)
    max_val = np.max(column)
    if max_val == 0:
        return 0
    return min_val / max_val

result = np.apply_along_axis(divide_min_max, axis=0, arr=arr)
print("min/max for columns:", result)
```

8) Ð’Ñ‹Ð¿Ð¾Ð»Ð½Ð¸Ñ‚ÑŒ Ñ€Ð°Ð½Ð¶Ð¸Ñ€Ð¾Ð²ÐºÑƒ ÑÐ»ÐµÐ¼ÐµÐ½Ñ‚Ð¾Ð² Ð¼Ð°ÑÑÐ¸Ð²Ð° NumPy.

ÐžÐ¶Ð¸Ð´Ð°ÐµÐ¼Ñ‹Ð¹ Ð²Ñ‹Ð²Ð¾Ð´ Ð´Ð»Ñ Ð¼Ð°ÑÑÐ¸Ð²Ð° [5 63 33 78 70 72 79 90 21 89]: [0 3 2 6 4 5 7 9 1 8].

```python
def rank_elements(arr):
    sorted_indices = np.empty_like(arr)
    sorted_indices[arr.argsort()] = np.arange(len(arr))
    return sorted_indices

arr = np.array([5, 63, 33, 78, 70, 72, 79, 90, 21, 89])
output = rank_elements(arr)
print("Ð Ð°Ð½Ð¶Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð½Ñ‹Ðµ ÑÐ»ÐµÐ¼ÐµÐ½Ñ‚Ñ‹: ", output)
```
