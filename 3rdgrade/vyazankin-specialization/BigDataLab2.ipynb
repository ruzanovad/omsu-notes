```python
import numpy as np
```

2) –°–æ–∑–¥–∞–π—Ç–µ –º–∞—Å—Å–∏–≤ a —Ä–∞–∑–º–µ—Ä–æ–º 5x2. –ü–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —á–∏—Å–ª–∞ 1, 2, 3, 3, 1, –∞ –≤—Ç–æ—Ä–æ–π - —á–∏—Å–ª–∞ 6, 8, 11, 10, 7.

–ë—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –∫–∞–∂–¥—ã–π —Å—Ç–æ–ª–±–µ—Ü - —ç—Ç–æ –ø—Ä–∏–∑–Ω–∞–∫, –∞ —Å—Ç—Ä–æ–∫–∞ - –Ω–∞–±–ª—é–¥–µ–Ω–∏–µ.

–ó–∞—Ç–µ–º –Ω–∞–π–¥–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ –∫–∞–∂–¥–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É. –†–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–ø–∏—à–∏—Ç–µ –≤ –º–∞—Å—Å–∏–≤ mean_a, –≤ –Ω–µ–º –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å 2 —ç–ª–µ–º–µ–Ω—Ç–∞.

```python
a = np.array([[1, 6],
              [2, 8],
              [3, 11],
              [3, 10],
              [1, 7]])

mean_a = np.mean(a, axis=0)

print("–°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ –∫–∞–∂–¥–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É: ", mean_a)
```

3) –í—ã—á–∏—Å–ª–∏—Ç–µ –º–∞—Å—Å–∏–≤ a_centered, –æ—Ç–Ω—è–≤ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏–π –º–∞—Å—Å–∏–≤–∞ –∞ –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ—Å—è –≤ –º–∞—Å—Å–∏–≤–µ mean_a.

–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–æ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—å—Å—è –≤ –æ–¥–Ω–æ –¥–µ–π—Å—Ç–≤–∏–µ.

```python
a_centered = a - mean_a

print("–ú–∞—Å—Å–∏–≤ a_centered:\n", a_centered)
```

4) –ù–∞–π–¥–∏—Ç–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤ –º–∞—Å—Å–∏–≤–∞ a_centered. –í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –¥–æ–ª–∂–Ω–∞ –ø–æ–ª—É—á–∏—Ç—å—Å—è –≤–µ–ª–∏—á–∏–Ω–∞ a_centered_sp. –ó–∞—Ç–µ–º –ø–æ–¥–µ–ª–∏—Ç–µ a_centered_sp –Ω–∞ N-1, –≥–¥–µ N - —á–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π.

```python
a_centered_sp = np.dot(a_centered[:, 0], a_centered[:, 1])
# –ò–ª–∏ a_centered[:, 0] @ a_centered[:, 1]

N = a_centered.shape[0]
result = a_centered_sp / (N - 1)

print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç 1: (a_centered_sp / (N-1)) = ({a_centered_sp} / {N - 1}) = {result}")

# –î–ª—è –∑–∞–¥–∞—á–∏ —Å —á–∏—Å–ª–æ–º —Å—Ç–æ–ª–±—Ü–æ–≤ > 2:
def scalar_product(data):
    selected_columns = np.array(data[:, 0])
    for i in range(1, data.shape[1]):
        selected_columns *= data[:, i]
    return np.sum(selected_columns)

result = scalar_product(a_centered) / (N - 1)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç 2: (scalar_product(a_centered) / (N-1)) = ({scalar_product(a_centered)} / {N - 1}) = {result}")
```

5) –í —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –ø–æ–ª—É—á–∏–≤—à–µ–µ—Å—è —á–∏—Å–ª–æ, –≤—ã—á–∏—Å–ª–∏–≤ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—é –µ—â–µ –æ–¥–Ω–∏–º —Å–ø–æ—Å–æ–±–æ–º - —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ np.cov. –í –∫–∞—á–µ—Å—Ç–≤–µ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ m —Ñ—É–Ω–∫—Ü–∏—è np.cov –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤ a. –í –ø–æ–ª—É—á–∏–≤—à–µ–π—Å—è –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ (–º–∞—Å—Å–∏–≤ Numpy —Ä–∞–∑–º–µ—Ä–æ–º 2x2) –∏—Å–∫–æ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ –±—É–¥–µ—Ç —Ä–∞–≤–Ω–æ —ç–ª–µ–º–µ–Ω—Ç—É –≤ —Å—Ç—Ä–æ–∫–µ —Å –∏–Ω–¥–µ–∫—Å–æ–º 0 –∏ —Å—Ç–æ–ª–±—Ü–µ —Å –∏–Ω–¥–µ–∫—Å–æ–º 1.

```python
cov_matrix = np.cov(a.T)
covariance_np = cov_matrix[0, 1]
print("–ö–æ–≤–∞—Ä–∏–∞—Ü–∏—è (np.cov):", covariance_np)
```

6) –ù–∞–π—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–µ–ª–µ–Ω–∏—è –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ –¥–≤—É–º–µ—Ä–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞, –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏—é np.apply_along_axis().

```python
arr = np.array([[4, -2, -3],
                [8,  3, -6],
                [1, 10,  0]])

def divide_min_max(column):
    min_val = np.min(column)
    max_val = np.max(column)
    if max_val == 0:
        return 0
    return min_val / max_val

result = np.apply_along_axis(divide_min_max, axis=0, arr=arr)
print("min/max for columns:", result)
```

7) –ù–∞–π—Ç–∏ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –∑–Ω–∞—á–µ–Ω–∏—è –≤ –º–∞—Å—Å–∏–≤–µ NumPy.

–î–ª—è —ç—Ç–æ–≥–æ –Ω–∞–π–¥–∏—Ç–µ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è
–∑–Ω–∞—á–µ–Ω–∏—è (–Ω–∞—á–∏–Ω–∞—è —Å–æ –≤—Ç–æ—Ä–æ–≥–æ –≤—Ö–æ–∂–¥–µ–Ω–∏—è) –≤ –∑–∞–¥–∞–Ω–Ω–æ–º –º–∞—Å—Å–∏–≤–µ –∏ –æ—Ç–º–µ—Ç—å—Ç–µ –∏—Ö –∫–∞–∫ True. –ü–µ—Ä–≤–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –æ—Ç–º–µ—á–∞–π—Ç–µ –∫–∞–∫ False.

–î–ª—è –º–∞—Å—Å–∏–≤–∞ [1 5 4 0 1 3 4 1 5 0] –æ–∂–∏–¥–∞–µ–º—ã–π –≤—ã–≤–æ–¥ [False False False False True False True True True True].

```python
arr = np.array([1, 5, 4, 0, 1, 3, 4, 1, 5, 0])

duplicates = np.full(arr.shape, False)
for i in range(len(arr)):
    if arr[i] in arr[:i]:
        duplicates[i] = True

print("–ú–µ—Ç–∫–∏ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏—Ö—Å—è –∑–Ω–∞—á–µ–Ω–∏–π: ", duplicates)

duplicates = np.full(arr.shape, True)
unique, indices = np.unique(arr, return_index=True)
duplicates[indices] = False

print("–ú–µ—Ç–∫–∏ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏—Ö—Å—è –∑–Ω–∞—á–µ–Ω–∏–π: ", duplicates)
```

8) –í—ã–ø–æ–ª–Ω–∏—Ç—å —Ä–∞–Ω–∂–∏—Ä–æ–≤–∫—É —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞ NumPy.

–û–∂–∏–¥–∞–µ–º—ã–π –≤—ã–≤–æ–¥ –¥–ª—è –º–∞—Å—Å–∏–≤–∞ [5 63 33 78 70 72 79 90 21 89]: [0 3 2 6 4 5 7 9 1 8].

```python
def rank_elements(arr):
    sorted_indices = np.empty_like(arr)
    sorted_indices[arr.argsort()] = np.arange(len(arr))
    return sorted_indices

arr = np.array([5, 63, 33, 78, 70, 72, 79, 90, 21, 89])
output = rank_elements(arr)
print("–†–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã: ", output)
```

–°–æ–∑–¥–∞–π—Ç–µ –º–∞—Ç—Ä–∏—Ü—ã ùê¥ –∏ ùêµ —Ä–∞–∑–º–µ—Ä–∞–º–∏ 3—Ö2 –∏ 2—Ö2 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ tr(ùê¥ùêµùêµA^T) –∏—Å–ø–æ–ª—å–∑—É—è —Ñ—É–Ω–∫—Ü–∏—é np.einsum().

```python
A = np.array([[0, 1],
              [2, 3],
              [4, 5]])

B = np.array([[0, 1],
              [2, 3]])

trace = np.einsum('ij,jk,kl,il->', A, B, B, A)

print("–°–ª–µ–¥ –º–∞—Ç—Ä–∏—Ü—ã ABBA^T: ", trace)
```
