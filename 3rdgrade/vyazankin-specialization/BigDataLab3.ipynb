1) Найдите евклидово расстояние между двумя Series a и b, не используя встроенную формулу.

```python
a = pd.Series([2, 4, 6, 8])
b = pd.Series([1, 3, 5, 7])

diff = a - b

squared_diff = diff**2

sum_squared_diff = np.sum(squared_diff)

euclidean_distance = np.sqrt(sum_squared_diff)

print("Евклидово расстояние: ", euclidean_distance)
```

2) Найдите максимально возможное абсолютное значение корреляции каждого
столбца с другими столбцами следующего датафрейма.

```python
np.random.seed(10)
df = pd.DataFrame(np.random.randint(1, 100, 16).reshape(4, 4), columns=list('efgh'), index=list('abcd'))
print(df)

corr_matrix = df.corr().abs()

np.fill_diagonal(corr_matrix.values, 0)

max_corr_values = corr_matrix.max()

print("Максимальные значения корреляции:")
print(max_corr_values)
```

3) Нормализуйте все столбцы приведенного ниже датафрейма путем вычитания
среднего значения столбца и деления на стандартное отклонение.
Сделайте так, чтобы все значения в датафрейме находились в диапазоне от 0 до 1.

```python
np.random.seed(10)
df = pd.DataFrame(np.random.randint(1, 100, 16).reshape(4, 4))

df_normalized = (df - df.mean()) / df.std()

df_normalized = (df_normalized - df_normalized.min()) / (df_normalized.max() - df_normalized.min())

print(df_normalized)
```

5) К датафрейму, приведенному ниже, добавьте два столбца nearest_row и dist
содержащих номер ближайшего по евклидовому расстоянию столбца и само это
расстояние.

```python
df = pd.DataFrame(np.random.randint(1, 100, 16).reshape(4, 4), columns=list('efgh'), index=list('abcd'))

distances = np.sqrt(np.sum(np.square(df.values[:, np.newaxis] - df.values), axis=2))

# Исключаем расстояния до самих себя
np.fill_diagonal(distances, np.inf)
# Получение номеров ближайших строк и расстояний
nearest_row = np.argmin(distances, axis=1)
dist = np.min(distances, axis=1)

df['nearest_row'] = nearest_row
df['dist'] = dist

print(df)
```

6) Создайте датафрейм authors со столбцами author_id и author_name, в которых
соответственно содержатся данные: [1, 2, 3] и ['Тургенев', 'Чехов', 'Островский'].

Затем создайте датафрейм book cо столбцами author_id, book_title и price,в
которых соответственно содержатся данные: [1, 1, 1, 2, 2, 3, 3], ['Отцы и дети',
'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза',
'Таланты и поклонники'], [450, 300, 350, 500, 450, 370, 290].

```python
authors = pd.DataFrame({'author_id': [1, 2, 3],
                        'author_name': ['Тургенев', 'Чехов', 'Островский']})

books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],
                      'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо',
                                     'Толстый и тонкий', 'Дама с собачкой', 'Гроза',
                                     'Таланты и поклонники'],
                      'price': [450, 300, 350, 500, 450, 370, 290]})

print("Датафрейм authors:")
print(authors)

print("\nДатафрейм books:")
print(books)
```

7) Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.

```python
authors_price = pd.merge(authors, books, on='author_id')
print(authors_price)
```

8) Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью
самыми дорогими книгами.

```python
top5 = authors_price.sort_values('price', ascending=False).head(5)
print(top5)
```

9) Создайте датафрейм authors_stat на основе информации из authors_price.
В датафрейме authors_stat должны быть четыре столбца: author_name, min_price,
max_price и mean_price, в которых должны содержаться соответственно имя
автора, минимальная, максимальная и средняя цена на книги этого автора.

```python
authors_stat = authors_price.groupby('author_name').agg({
    'price': ['min', 'max', 'mean']
}).reset_index()

authors_stat.columns = ['author_name', 'min_price', 'max_price', 'mean_price']

print(authors_stat)
```

10) Создайте новый столбец в датафрейме authors_price под названием cover, в нем
будут располагаться данные о том, какая обложка у данной книги - твердая или

'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']. Просмотрите
документацию по функции pd.pivot_table. Для каждого автора посчитайте
суммарную стоимость книг в твердой и мягкой обложке. Используйте для этого
функцию pd.pivot_table. При этом столбцы должны называться "твердая" и
"мягкая",а индексами должны быть фамилии авторов. Пропущенные значения
стоимостей заполните нулями. Назовите полученный датасет book_info и
сохраните его в формат pickle под названием "book_info.pkl". Затем загрузите из

```python
cover_data = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price['cover'] = cover_data

book_info = pd.pivot_table(authors_price, values='price', index='author_name', columns='cover', aggfunc='sum', fill_value=0)
book_info.columns = ['твердая', 'мягкая']
#print(book_info)

book_info.to_pickle('book_info.pkl')
book_info2 = pd.read_pickle('book_info.pkl')

are_equal = book_info.equals(book_info2)
print(are_equal)
```

который по названию штата необходимо заполнить аббревиатурой.

```python
!pip install fuzzywuzzy
from fuzzywuzzy import process

state_abbreviations = {
    "VERMONT": "VT", "GEORGIA": "GA", "IOWA": "IA", "Armed Forces Pacific": "AP",
    "GUAM": "GU", "KANSAS": "KS", "FLORIDA": "FL", "AMERICAN SAMOA": "AS",
    "NORTH CAROLINA": "NC", "HAWAII": "HI", "NEW YORK": "NY", "CALIFORNIA": "CA",
    "ALABAMA": "AL", "IDAHO": "ID", "FEDERATED STATES OF MICRONESIA": "FM",
    "Armed Forces Americas": "AA", "DELAWARE": "DE", "ALASKA": "AK", "ILLINOIS": "IL",
    "Armed Forces Africa": "AE", "SOUTH DAKOTA": "SD", "CONNECTICUT": "CT", "MONTANA": "MT",
    "MASSACHUSETTS": "MA", "PUERTO RICO": "PR", "Armed Forces Canada": "AE", "NEW HAMPSHIRE": "NH",
    "MARYLAND": "MD", "NEW MEXICO": "NM", "MISSISSIPPI": "MS", "TENNESSEE": "TN",
    "PALAU": "PW", "COLORADO": "CO", "Armed Forces Middle East": "AE", "NEW JERSEY": "NJ",
    "UTAH": "UT", "MICHIGAN": "MI", "WEST VIRGINIA": "WV", "WASHINGTON": "WA",
    "MINNESOTA": "MN", "OREGON": "OR", "VIRGINIA": "VA", "VIRGIN ISLANDS": "VI",
    "MARSHALL ISLANDS": "MH", "WYOMING": "WY", "OHIO": "OH", "SOUTH CAROLINA": "SC",
    "INDIANA": "IN", "NEVADA": "NV", "LOUISIANA": "LA", "NORTHERN MARIANA ISLANDS": "MP",
    "NEBRASKA": "NE", "ARIZONA": "AZ", "WISCONSIN": "WI", "NORTH DAKOTA": "ND",
    "Armed Forces Europe": "AE", "PENNSYLVANIA": "PA", "OKLAHOMA": "OK", "KENTUCKY": "KY",
    "RHODE ISLAND": "RI", "DISTRICT OF COLUMBIA": "DC", "ARKANSAS": "AR", "MISSOURI": "MO",
    "TEXAS": "TX", "MAINE": "ME"
}

# Функция для сопоставления имен штатов с аббревиатурами
def match_state(state_name):
    # 94 - максимальное значение при котором сопоставляются все штаты
    match = process.extractOne(state_name, state_abbreviations.keys(), score_cutoff=94)
    if match:
        return state_abbreviations[match[0]]
    return None

# Добавление нового столбца abbrev с аббревиатурами штатов
abbrev = df['state'].apply(match_state)
df.insert(df.columns.get_loc("postal-code") + 1, "abbrev", abbrev)
df
```

14) Выполнив соответствующую группировку, подсчитайте объем продаж по каждому
штату.

```python
import locale

# Устанавливаем локаль для форматирования в долларах
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
def format_dollars(number):
    return locale.currency(number, grouping=True, symbol=True).split(".")[0]

grouped_data = df.groupby('abbrev')[['Jan', 'Feb', 'Mar', 'total']].sum()
grouped_data.applymap(format_dollars)
```
